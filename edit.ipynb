```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

cer_crops = pd.read_excel('cereal_crops.xlsx')
vegetables = pd.read_excel('vegetables.xlsx')
fr_spices = pd.read_excel('fruits_spices.xlsx')

```

```python
#remove the values before "-" in the second column
vegetables['Product Name'] = vegetables['Product Name'].str.split('-').str[1]
#remove spaces in the second column
vegetables['Product Name'] = vegetables['Product Name'].str.strip()
#save the excel file
vegetables.to_excel('vegetabless.xlsx', index=False)
```

```python
#remove the values before "-" in the second column of fruits_spices
fr_spices['Product Name'] = fr_spices['Product Name'].str.split('-').str[1]
#remove spaces in the second column
fr_spices['Product Name'] = fr_spices['Product Name'].str.strip()
#save the excel file
fr_spices.to_excel('fruits_spicess.xlsx', index=False)
```

```python
#load data.xlsx
data = pd.read_excel('data.xlsx')
#save the excel file
data.to_excel('all_data.xlsx', index=False)
```

```python

df = pd.read_excel('data.xlsx')
## all the numerical data is in Tonnes
df.head()
#group the data by product category, sort it and plot horizontal bar chart
#change x label
df.groupby('Product Category').sum().sort_values(by=2020).iloc[3:-8].plot(kind='barh', figsize=(20,20), ylabel='Tonnes')
#plot the last 6 rows of the data on a different chart
df.groupby('Product Category').sum().sort_values(by=2020).iloc[-8:].plot(kind='barh', figsize=(20,20), ylabel='Tonnes')
df.groupby('Product Category').sum().sort_values(by=2020).iloc[0:3].plot(kind='barh', figsize=(20,20), ylabel='Tonnes')
```
