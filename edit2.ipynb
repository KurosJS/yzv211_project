```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#getting database
df = pd.read_excel('data2.xlsx')
df.head()
```

```python
#convert the data to numeric
df['2002'] = pd.to_numeric(df['2002'], errors='coerce')
#convert the data to numeric
df['2011'] = pd.to_numeric(df['2011'], errors='coerce')
#convert the data to numeric
df['2012'] = pd.to_numeric(df['2012'], errors='coerce')
#group the data by Category, sort it and plot horizontal bar chart
df.groupby('Category')['2002','2011','2012'].sum().sort_values(by='2012', ascending=True).plot(kind='barh', figsize=(10,10))

#plot the last 3 rows of the data on a different chart by grouping the data by Category, sort it and plot horizontal bar chart
df.groupby('Category')['2002','2011','2012'].sum().sort_values(by='2012', ascending=False).tail(3).plot(kind='barh', figsize=(10,10))
```
